```python
import librosa
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
from IPython.display import Audio
import librosa.display
```

```python
y, sr = librosa.load("../Data/White Noise.wav", sr=None)
```

```python
def gaussian_smoothing(x, sigma=2):
    return gaussian_filter1d(x, sigma)
```

```python
gaussina_filter_audio = gaussian_filter1d(y, sigma=2)
```

```python
plt.figure(figsize=(10,2.5))
librosa.display.waveshow(y, sr=sr, alpha=0.8)
plt.title("White Noise")
plt.tight_layout()
plt.show()
display(Audio(data=y, rate=sr))
```

```python
plt.figure(figsize=(10,2.5))
librosa.display.waveshow(gaussina_filter_audio, sr=sr, alpha=0.8)
plt.title("White Noise")
plt.tight_layout()
plt.show()
display(Audio(data=gaussina_filter_audio, rate=sr))
```
