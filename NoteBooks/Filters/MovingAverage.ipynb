```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import soundfile as sf
from scipy.signal import convolve
from scipy.ndimage import gaussian_filter1d
from IPython.display import Audio
import librosa.display
```

```python
y, sr = librosa.load("../../Data/Noise/White Noise.wav", sr=None)
```

```python
def moving_average(x, window_size=5):
    kernel = np.ones(window_size) / window_size
    return convolve(x, kernel, mode='same')
```

```python
moving_average_audio = moving_average(y)
```

```python
plt.figure(figsize=(10,2.5))
librosa.display.waveshow(moving_average_audio, sr=sr, alpha=0.8)
plt.title("Moving Average on White Noise")
plt.tight_layout()
plt.show()
display(Audio(data=moving_average_audio, rate=sr))
```

```python
plt.figure(figsize=(10,2.5))
librosa.display.waveshow(y, sr=sr, alpha=0.8)
plt.title("White Noise")
plt.tight_layout()
plt.show()
display(Audio(data=y, rate=sr))
```
